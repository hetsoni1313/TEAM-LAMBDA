```python
suspect = []
for c in df.columns:
    if df[c].dtype == object:
        sample = df[c].dropna().astype(str).head(30)
        numeric_like = sample.str.replace(",", "", regex=False).str.isnumeric().mean()
        if numeric_like > 0.5:
            suspect.append(c)

print(suspect)
```

```python
financial_cols = [
    "Number of All Returns",
    "NY AGI of All Returns (in thousands) *",
    "Deductions of All Returns (in thousands)",
    "Dependent Exemptions of All Returns (in thousands)",
    "Taxable Income of All Returns (in thousands)",
    "Tax Before Credits of All Returns (in thousands)",
    "Tax Liability of All Returns (in thousands) **"
]
```

```python
df["financial_missing"] = df[financial_cols].isna().sum(axis=1)
df = df[df["financial_missing"] < len(financial_cols)]
```

```python
df = df.drop(columns=["financial_missing"])
```

```python
df.shape
```

```python
def parse_income_range(x):
    s = str(x)
    if "Under" in s:
        num = ''.join([ch for ch in s if ch.isdigit()])
        return 0, int(num), int(num) / 2
    if "-" in s:
        lo, hi = s.split("-")
        lo = int(lo.replace(",", "").strip())
        hi = int(hi.replace(",", "").strip())
        return lo, hi, (lo + hi) / 2
    return np.nan, np.nan, np.nan

df["Income_Min"], df["Income_Max"], df["Income_Mid"] = zip(*df["Income Class"].apply(parse_income_range))
```

```python
df = df.drop(columns=["Place of Residence Sort Order", "Income Class Sort Order"])
```

```python
for col in financial_cols:
    df[col] = df[col].clip(lower=0)
```

```python
drop_cols = [
    "Resident Type", "Place of Residence", "Country",
    "State", "County", "Income Class", "Disclosure",
]
```

```python
df2 = df.drop(columns=drop_cols)
```

```python
df2 = df2.dropna(subset=["Income_Mid"]).reset_index(drop=True)
```

```python
numeric_features = [
    "Number of All Returns",
    "NY AGI of All Returns (in thousands) *",
    "Deductions of All Returns (in thousands)",
    "Dependent Exemptions of All Returns (in thousands)",
    "Taxable Income of All Returns (in thousands)",
    "Tax Before Credits of All Returns (in thousands)",
    "Income_Mid"
]
```

# STEP-2 MODEL

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
```

```python
X = df2[numeric_features].values
y = df2[target].values * 1000
```

```python
y_log = np.log1p(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_log, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
params = {
    'n_estimators': [150, 300],
    'learning_rate': [0.03, 0.05],
    'max_depth': [2, 3],
    'min_samples_split': [20, 40],
    'min_samples_leaf': [5, 10],
    'subsample': [0.7, 0.9]
}

gb = GradientBoostingRegressor(random_state=42)

grid = GridSearchCV(
    estimator=gb,
    param_grid=params,
    cv=4,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

grid.fit(X_train_s, y_train)

best_model = grid.best_estimator_
```

```python
grid.best_params_
```

```python
train_pred = np.expm1(best_model.predict(X_train_s))
test_pred = np.expm1(best_model.predict(X_test_s))

train_y = np.expm1(y_train)
test_y = np.expm1(y_test)

train_rmse = np.sqrt(mean_squared_error(train_y, train_pred))
test_rmse = np.sqrt(mean_squared_error(test_y, test_pred))

train_r2 = r2_score(train_y, train_pred)
test_r2 = r2_score(test_y, test_pred)
```

```python
print("Train RMSE:", train_rmse)
print("Train RÂ²:", train_r2)

print("Test RMSE:", test_rmse)
print("Test RÂ²:", test_r2)
```

```python
print("RMSE Gap:", abs(train_rmse - test_rmse))
print("RÂ² Gap:", abs(train_r2 - test_r2))

if abs(train_r2 - test_r2) < 0.05 and abs(train_rmse - test_rmse) < 0.15 * test_rmse:
    print("Model is NOT Overfitting")
else:
    print(" Model is Overfitting")
```

```python
plt.figure(figsize=(7,6))
plt.scatter(test_y, test_pred, alpha=0.5)
plt.plot([test_y.min(), test_y.max()],
         [test_y.min(), test_y.max()], 'r--')
plt.xlabel("Actual Tax Liability ($)")
plt.ylabel("Predicted Tax Liability ($)")
plt.title("Actual vs Predicted Tax Liability (Final Model)")
plt.show()
```
